```python
group_counts = np.bincount(group_sizes)
```

```python
len(group_counts)
```

```python
counts = []
sizes = []
for i, cnt in enumerate(group_counts):
    if cnt > 0:
        counts.append(cnt)
        sizes.append(i)
```

```python
plt.figure(figsize=(15, 7))
plt.scatter(np.log2(sizes), np.log2(counts))
plt.xlabel('log_2(group size)', fontsize=15)
plt.ylabel('log_2(group count)', fontsize=15)
plt.xticks(np.arange(0, 10, 1))
plt.grid(ls=':')
plt.show()
```
