Реализовывал предложенный подход с разделением хэшей на 4 равные части. Комментарии есть в коде

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('output.txt', 'r') as rf:
    mas = [int(s) for s in rf.readlines()]
```

Проверим, корректно ли отработала программа

```python
sum(mas)
```

```python
min(mas)
```

Вроде ок. Теперь построим соответствующий график

```python
x = []
y = []
for i, val in enumerate(np.bincount(mas)):
    if val is not 0:
        x.append(np.log(i))
        y.append(np.log(val))
```

```python
plt.figure(figsize=(16, 8))
plt.xlabel('log(group size)')
plt.ylabel('log(count)')
plt.scatter(x, y)
```
